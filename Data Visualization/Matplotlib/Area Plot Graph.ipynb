### Multiple Areas

```python
a = [1,2,3,4,5]
area1 = [2,3,2,4,3]
area2 = [2,3,5,4,7]
plt.stackplot(a, area1, area2, colors=["red","yellow"])
plt.show()
```

```python
a = [1,2,3,4,5]
area1 = [2,3,2,4,3]
area2 = [2,3,5,4,7]
area3 = [2,3,4,3,5]
plt.stackplot(a, area1, area2,area3, colors=["red","yellow", 'grey'])
plt.show()
```

```python
a = [1,2,3,4,5]
area1 = [6,3,5,4,2]
area2 = [2,3,5,4,7]
area3 = [2,3,4,3,5]
plt.stackplot(a, area1, area2,area3, colors=["red","yellow", 'grey'], labels=["Area1", "Area2", "Area3"])
plt.legend(loc = 2)
plt.show()
```

```python
a = [1,2,3,4,5]
area1 = [6,3,5,4,2]
area2 = [2,3,5,4,7]
area3 = [2,3,4,3,5]
area4 = [3,4,5,6,3]
plt.stackplot(a, area1, area2,area3,area4, colors=["red","yellow", 'grey','orange'], 
              labels=["Area1", "Area2", "Area3",'Area4'], baseline='sym')
plt.legend(loc = 2)
plt.show()
```

```python
a = [1,2,3,4,5]
area1 = [6,3,5,4,2]
area2 = [2,3,5,4,7]
area3 = [2,3,4,3,5]
area4 = [3,4,5,6,3]
plt.stackplot(a, area1, area2,area3,area4, colors=["red","yellow", 'grey','orange'], 
              labels=["Area1", "Area2", "Area3",'Area4'], baseline='wiggle')
plt.legend(loc = 2)
plt.show()
```

```python
a = [1,2,3,4,5]
area1 = [6,3,5,4,2]
area2 = [2,3,5,4,7]
area3 = [2,3,4,3,5]
area4 = [3,4,5,6,3]
plt.stackplot(a, area1, area2,area3,area4, colors=["red","yellow", 'grey','orange'], 
              labels=["Area1", "Area2", "Area3",'Area4'], baseline='weighted_wiggle')
plt.legend(loc = 2)
plt.show()
```
