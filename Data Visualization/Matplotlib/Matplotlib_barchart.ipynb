###  A bar graph is a pictorial representation of data, quantities, or numbers using bars, columns, or strips

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
fruits = np.array(["APPLE","ORANGE","BANANA","MANGO","KIWI"])
```

```python
numbers = np.array([150,220,180,200,120])
```

```python
plt.bar(fruits, numbers)
```

```python
plt.bar(fruits, numbers, color = "r")
```

```python
plt.barh(fruits, numbers)
```

```python
plt.bar(fruits, numbers, 0.4)
```

```python
plt.barh(fruits, numbers, color = 'skyblue')
```

```python
plt.barh(fruits, numbers, 0.4, color = 'skyblue') 
```

```python
plt.bar(fruits, numbers, color = 'pink')
plt.title("FRUITS QUANTITY")
plt.xlabel("FRUITS")
plt.ylabel("QUANTITY")
plt.show()
```

```python
plt.style.use("default")
plt.bar(fruits, numbers, color = 'pink', linewidth = 5, edgecolor = "red")
plt.title("FRUITS QUANTITY")
plt.xlabel("FRUITS")
plt.ylabel("QUANTITY")
plt.show()
```

```python
plt.style.use('ggplot')
plt.bar(fruits, numbers, color = 'grey', linewidth = 3, edgecolor = 'red')
plt.title("FRUITS QUANTITY")
plt.xlabel("FRUITS")
plt.ylabel("QUANTITY")
plt.show()
```

### STACKED GRAPH

```python
import numpy as np
```

```python
clist = ['USA', 'JAPAN', 'INDIA', 'RUSSIA','FRANCE', 'ISRAEL']
```

```python
gold = np.random.randint(10,50,(6,))
```

```python
gold
```

```python
silver = np.random.randint(50,110,(6,))
```

```python
silver
```

```python
bronze = np.random.randint(150,300,(6,))
```

```python
bronze
```

```python
medals = gold + silver + bronze
```

```python
medals
```

```python
plt.style.use("default")
plt.bar(clist, medals, 0.6, color = "skyblue", linewidth = 2 , edgecolor = "brown")
plt.legend(["MEDALS"])
plt.title("MEDALS TALLY")
plt.xlabel("COUNTRY")
plt.ylabel("MEDALS")
plt.show()
```

```python
plt.bar(clist, bronze,0.4, color = "red", label = "Bronze")
plt.bar(clist,silver,0.4, color = "blue", label = "Silver")
plt.bar(clist, gold, 0.4, color = "yellow", label = "Gold")
plt.legend()
```

```python
plt.bar(clist, bronze,0.8, color = "grey", label = "Bronze")
plt.bar(clist,silver,0.6, color = "blue", label = "Silver")
plt.bar(clist, gold, 0.4, color = "yellow", label = "Gold")
plt.legend()
```

### Stacked Bar Graph

```python
plt.bar(clist, gold, 0.5, color = "yellow")
plt.bar(clist, silver, 0.5, bottom=gold,color = 'red')
plt.bar(clist,bronze, 0.5, bottom = silver, color = 'blue')
```

```python
course = ["DS", "DE", "DA", "PY", "DL"]
# courseno = [0,1,2,3,4]
courseno1 = np.arange(len(course))
courseno2 = courseno1 + 0.4
plt.bar(courseno1, boys, 0.4, color = "orange", label = "Boys")
plt.bar(courseno2, girls, 0.4, color = "pink", label = "girls")
plt.legend()
```

```python
course = ["DS", "DE", "DA", "PY", "DL"]
# courseno = [0,1,2,3,4]
courseno1 = np.arange(len(course))
courseno2 = courseno1 + 0.4
plt.bar(courseno1, boys, 0.4, color = "orange", label = "Boys", linewidth = 3, edgecolor = "red")
plt.bar(courseno2, girls, 0.4, color = "pink", label = "girls",linewidth = 3, edgecolor = "red")
plt.legend()
```
