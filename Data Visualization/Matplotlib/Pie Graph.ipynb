```python
import matplotlib.pyplot as plt
```

```python
smedia = ['Youtube', 'Instagram', 'Twitter','Facebook', 'whatsapp']
```

```python
stds = [40, 35, 25, 60, 50]
```

```python
plt.pie([1])
plt.show()
```

```python
smedia
```

```python
stds
```

```python
plt.pie(stds)
plt.show()
```

```python
plt.pie(stds, labels = smedia)
plt.show()
```

```python
smedia
```

```python
e = [0, 0, 0, 0.05, 0]
```

```python
e
```

```python
plt.pie(stds, labels=smedia, explode=e)
plt.show()
```

```python
smedia
```

```python
e = [0.1, 0, 0, 0.1, 0]
```

```python
plt.pie(stds, labels=smedia, explode=e)
plt.show()
```

```python
smedia
```

```python
c = ['r','y','g','b','m']
```

```python
plt.pie(stds, labels=smedia, explode=e, colors= c)
plt.show()
```

```python
plt.pie(stds, labels=smedia, explode=e, colors= c, autopct="%0.2F%%")
plt.show()
```

```python
plt.pie(stds, labels=smedia, explode=e, colors= c, autopct="%0.2F%%",shadow=True)
plt.show()
```

```python
plt.pie(stds, labels=smedia, explode=e, colors= c, autopct="%0.2F%%",shadow=True,radius = 1.4)
plt.show()
```
