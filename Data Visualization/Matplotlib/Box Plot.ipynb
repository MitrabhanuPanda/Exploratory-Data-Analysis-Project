```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50]

plt.boxplot(x)
plt.show()
```

```python
x = [10,20,30,40,50]
plt.boxplot(x, notch=True)
plt.show()
```

```python
x = [10,20,30,40,50]
plt.boxplot(x)
plt.show()
```

```python
x = [10,20,30,40,50]
plt.boxplot(x, vert=False)
plt.show()
```

```python
x = [10,20,30,40,50]
plt.boxplot(x, vert= False, widths= 0.5)
plt.show()
```

```python
x = [10,20,30,40,50]
plt.boxplot(x,)
plt.show()
```

```python
x = [10,20,30,40,50]
plt.boxplot(x, widths= 0.5)
plt.show()
```

```python
x = [10,20,30,40,50]
plt.boxplot(x, labels=["Python"])
plt.show()
```

```python
x = [10,20,30,40,50]
plt.boxplot(x, labels=["Python"], patch_artist= True)
plt.show()
```

```python
x = [10,20,30,40,50]
plt.boxplot(x, labels=["Python"], patch_artist= False)
plt.show()
```

```python
x = [10,20,30,40,50]
plt.boxplot(x, labels=["Python"], showmeans=True)
plt.show()
```

```python
x = [10,20,30,40,50, 140, 250]
plt.boxplot(x, labels=["Python"], showmeans=True)
plt.show()
```

```python
x = [-150,-50, 10,20,30,40,50, 140, 250]
plt.boxplot(x, labels=["Python"], showmeans=True, whis=2.5)
plt.show()
```

```python
x = [-150,-50, 10,20,30,40,50, 150]
plt.boxplot(x, labels=["Python"], showmeans=True)
plt.show()
```

```python
x = [-150,-50, 10,20,30,40,50, 150]
plt.boxplot(x, labels=["Python"], showmeans=True,sym='g+' )
plt.show()
```

```python
x = [10,20,30,40,50]
x1 = [10,20,30,40,50]
plt.boxplot(x, labels=["Python"], showmeans=True)
plt.boxplot(x1, labels=["Java"], showmeans=True)

plt.show()
```

```python
x = [10,20,30,40,50]
x1 = [10,20,30,40,50]
z = [x,x1]
plt.boxplot(z, labels=["Java","Python"], showmeans=True)

plt.show()
```
