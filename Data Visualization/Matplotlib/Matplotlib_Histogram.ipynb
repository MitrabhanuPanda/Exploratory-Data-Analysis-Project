```python
female_age = np.random.randint(18, 50,(50,))
```

```python
male_age = np.random.randint(20,60,(50,))
```

```python
male_age
```

```python
plt.hist([male_age], bins = 20, rwidth=0.90, color ="red", ec = "yellow")
plt.show()
```

```python
plt.hist(female_age, bins = 20, rwidth = 0.90, color = 'pink', ec = 'blue')
plt.show()
```

```python
plt.hist([male_age, female_age], bins=20, rwidth=0.90)
plt.show()
```

```python
plt.hist([male_age, female_age], bins=20, rwidth=0.90, color = ['red','pink'])
plt.show()
```

```python
plt.hist([male_age, female_age], bins=20, rwidth=0.90, orientation = "horizontal",color = ['red','pink'], ec = "yellow")
plt.show()
```

```python
plt.hist([male_age, female_age], bins=20, rwidth=0.90, color=['yellow','pink'])
plt.legend(["Male",'Female'])
plt.show()
```

```python
plt.hist([male_age, female_age], bins=20, rwidth=0.90, color=['yellow','pink'], label=["Male", 'Female'])
plt.legend()
plt.show()
```

```python
plt.hist([male_age], bins=20, rwidth=0.90, color=['yellow'], label=["Male"])
plt.hist([female_age], bins=20, rwidth=0.90, color=['red'], label=["Female"])

plt.legend()
plt.show()
```

```python
plt.hist([male_age, female_age], bins=20, rwidth=0.90, color=['yellow','pink'], label=["Male", 'Female'])
plt.legend()
plt.show()
```

```python
plt.hist([male_age], bins=20, rwidth=0.90, color=['yellow'], label=["Male"],orientation="horizontal")
plt.hist([female_age], bins=20, rwidth=0.90, color=['red'], label=["Female"], orientation='vertical')

plt.legend()
plt.show()
```
