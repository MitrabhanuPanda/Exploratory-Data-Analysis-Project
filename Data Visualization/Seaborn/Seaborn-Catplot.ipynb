```python
df_loan=pd.read_csv("loandata.csv")
df_sales = pd.read_csv("Sales_Data.csv")
```

```python
sns.catplot(x="Gender", y="ApplicantIncome", data=df_loan)
plt.show()
```

Changing the order of categories

In our plot, the quality of the cut is from best to worst. But let’s reverse the order: 

```python
df_loan["Property_Area"].unique()
```

```python
sns.catplot(x='Property_Area', data=df_loan, kind='count')
plt.show()
```

```python
category_order = ['Rural', 'Urban', 'Semiurban']

sns.catplot(x='Property_Area', data=df_loan, kind='count', order=category_order);
plt.show()
```

It is also important to keep in mind that a bar plot shows only the mean (or other estimator) value as default

```python
sns.catplot(x='Item_Type',
            y='Total_Profit',
            data=df_sales,
            kind='bar',
            ci=None)
plt.xticks(rotation = 90 )
plt.show()
```

```python
sns.catplot(x='Item_Type', y='Total_Profit', data=df_sales, kind='box')  
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.catplot(x='Sales_Channel', y='Total_Profit',data=df_sales,kind="violin")
plt.show()
```
