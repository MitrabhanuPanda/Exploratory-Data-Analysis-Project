```python
dfair = sns.load_dataset("flights")
```

```python
dfair
```

```python
flight = dfair.pivot_table( index = "year", columns = "month", values = "passengers", aggfunc="sum")
```

```python
sns.heatmap(flight)
```

```python
sns.heatmap(flight, linewidths=1 )
```

```python
sns.heatmap(flight, linewidths=1, linecolor="y")
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.heatmap(flight, linewidths=1, linecolor="y", annot=True)
```

```python
sns.heatmap(flight, linewidths=1, linecolor="y", annot=True, fmt="d")
plt.xticks(rotation = 45)
plt.show()
```

```python
sns.heatmap(flight, linewidths=1, linecolor="y", annot=True, fmt="d", cmap="summer")
plt.xticks(rotation = 45)
plt.show()
```

```python
sns.heatmap(flight, linewidths=1, linecolor="y", annot=True, fmt="d", cmap="winter")
plt.xticks(rotation = 45)
plt.show()
```

```python
sns.heatmap(flight, linewidths=1, linecolor="y", annot=True, fmt="d", cmap="xyz")
plt.xticks(rotation = 45)
plt.show()
```

```python
sns.heatmap(flight, linewidths=1, linecolor="y", annot=True, fmt="d", cmap="spring")
plt.xticks(rotation = 45)
plt.show()
```

```python
sns.heatmap(flight, linewidths=1, linecolor="y", annot=True, fmt="d", cmap="spring", cbar=False)
plt.xticks(rotation = 45)
plt.show()
```

```python
sns.heatmap(flight, linewidths=1, linecolor="y", annot=True, fmt="d", cmap="spring", cbar=False)

plt.xticks(rotation = 45)
plt.savefig("abc.jpeg")
plt.show()
```
