```python
dftips = pd.read_csv("tips.csv")
```

```python
sns.catplot(x = "tip", y = "size", data = dftips)
plt.show()
```

```python
dftip
```

```python
sns.catplot(x = "tip", y = "size", data = dftip, hue = "Gender")
plt.show()
```

```python
 sns.catplot(x = "tip", y = "size", data = dftip, hue = "Gender", palette='ocean_r')
plt.show()
```

```python
 sns.catplot(x = "tip", y = "size", data = dftip, hue = "Gender", palette='ocean_r', height=3)
plt.show()
```

```python
 sns.catplot(x = "tip", y = "size", data = dftip, hue = "Gender", palette='ocean_r', height=6)
plt.show()
```

```python
 sns.catplot(x = "tip", y = "size", data = dftip, hue = "Gender", palette='ocean_r', kind = "bar")
plt.show()
```

```python
 sns.catplot(x = "tip",data = dftip, hue = "Gender", palette='ocean_r', kind = "count")
plt.show()
```
