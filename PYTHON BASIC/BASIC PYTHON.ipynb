#### <font color="brown">7. BITWISE OPERATOR
* In python Bitwise Operators are used to perform Bitwise Calculation on Integers.
    
    * The Integers are 1st converted into binary & then operations are performed on bit by bit, hence the name bitwise operators. Then the result is returned in decimal format.
    
* It works only on __`INTEGER`__.

```python
s=10
w=30
```

```python
print(s&w)  # BITWISE AND OPERATOR
```

```python
print(s&w,bin(s&w))#bin is used to convert the integer to binary.
```

```python
print(s|w,bin(s|w))   # BITWISE OR OPERATOR
```

<font color="teal">BITWISE NOT / COMPLIMENT :</font> <br>

        * It is a Uninary Value.
        * Two Steps Process:
            * One's Compliment :-- Add one(1) to original Value.
            * Two's Compliment :-- Reverse the Sign.

```python
print(s^w,bin(s^w))  # BITWISE NOT OPERATOR
```

<font color="teal">RIGHT-SHIFT OPERATOR:</font><br>
    
    * Removes plays holder from right side of original value. 
    * Resultant is always less than original value.
    
<font color="teal"> LEFT-SHIFT OPERATOR:</font><br>
    
    * It is Opposite of Right Shift Bitwise Operator

```python
bin(s)
```

```python
s>>2  # BITWISE RIGHT SHIFT
```

```python
print(0b10)
```

```python
s<<2  # BITWISE LEFT SHIFT
```

    * CONTROL FLOW STATEMENT CONTROLS THE FLOW OF EXECUTION DURING EXECUTION PHASE.

#### <font color="red">WHAT IS ITERATION ?
* ITERATION IS A GENERAL TERM FOR TAKING EACH ITEM OF SOMETHING, ONE AFTER ANOTHER. ANY TIME YOU USE A LOOP, EXPLICIT OR IMPLICIT, TO GO OVER A GROUP OF ITEMS, THAT IS ITERATION.

* IN SIMPLE LANGUAGE ---> IT IS A PROCESS OF REPEATEDLY EXECUTING A BLOCK OF CODE.

```python
# EXAMPLE 
n=[1,2,3]

for i in n:
    print(i)
```

#### <font color="red">WHAT IS ITERATOR ?
* AN ITERATOR IS AN OBJECT THAT ALLOWS THE PROGRAMMER TO TRAVERSE THROUGH A SEQUENCE OF DATA WITHOUT HAVING TO STORE THE ENTIRE DATA IN THE MEMORY.
* IN SIMPLE LANGUAGE :---> IT IS AN OBJECT WHICH ACTS LIKE AN AGENT IN EXTRACTING DATA FROM INTERNAL MEMORY LOCATION OF SEQUENTIAL DATA TYPE.

```python
# EXAMPLE
l=[x  for x in range(1,10000)]

for i in l:
    print(i)    
```

```python
import sys
```

```python
sys.getsizeof(l)/64 # IT GIVE THE MEMORY OCCUPY BY THE LIST IN byte.
```

#### <font color="red">WHAT IS ITERABLE ?
* ITERABLE IS AN OBJECT WHICH ONE CAN ITERATE OVER.
* IN SIMPLE LANGUAGE ---> AN OBJECT ON WHICH A PROCESS OF ITERATION IS PERFORM.
* IT ARE SEQUENTIAL DATA TYPE (LIST, STRING, TUPLE)
* IT GENERATES AN ITERATOR WHEN PASSED TO __`iter()`__ METHOD.

```python
# EXAMPLE

m=[1,2,3]
print(type(m))

print()

# LIST IS AN ITERABLE.
print(type(iter(m)))

# iter(m) is ITERATOR.
```

### <font color="dark blue">__`range()`__ FUNCTION
* IT GENERATES THE VALUES CONSIDERING STARTING VALUE, ENDING VALUE AND STEP.

> * __STARING VALUE__ : FROM WHERE VALUES HAS TO BE GENERATED.

> * __STEP__ : IT DEFINES HOW TO GENERATES VALUE USING STARTING & ENDING VALUE.

### <font color="blue" size=5> b. INTERRUPTIVE STATEMENT
* IT INTERRUPTS THE FLOW OF STATEMENT.

> <b><font color="Purple" size=4>TYPES:<b/>
>> 1. BREAK
>> 2. CONTINUE
>> 3. EXIT
>> 4. PASS

### <font color="dark blue"> 1. BREAK
* IT IS USED TO STOP THE EXECUTION OF THE LOOP.
* AFTER BREAK STATEMENT.. EXECUTION CONTINUES WITH REST OF THE PROGRAM.

```python
for i in range(1,10): #break
    if i==7:
        break
        
    print(i)
    time.sleep(0.5)
print("This is the end of Program")
```

### <font color="dark blue"> 2. CONTINUE
* IT IS USED TO SKIP CURRENT ITERATION BASED ON CONDITION.

```python
for i in range(1,10): 
    if i==7:
        continue
    else:
        print(i)
        time.sleep(0.5)
```

### <font color="dark blue"> 3. PASS
* IT HELPS US TO EXECUTION THE PROGRAMM WITHOUT ANY NEW BLOCK IMPLEMENTATION.

```python
for i in range(1,10):
    pass
print("This is a Cat")
```

### <font color="dark blue"> 4. EXIT
* IT STOPS THE EXECUTION OF THE PROGRAM & TERMINATE COMPLETE PROGRAM.

```python
# WHEN YOU WILL EXECUTE THIS CODE,THEN THE KERNEL IS DEAD.
for i in range(1,10): 
    if i==7:
        exit(0) 
    
        print(i)
        time.sleep(0.5)
print("This is end of the Program")
```
