#### <font color="green">6. SPLIT
- It divide the String values by delimeter and store it into a `list`.

    `the sky --> ["the","sky"]` <br>
     here the delimeter is `space`.

#### <font color="red">REAL TIME EXAMPLE-1
- We want the Sure name from "Name" Column

```python
tit["Name"]
```

```python
# Split by the delimeter ","
tit["Name"].str.split(",")
```

```python
# By Using "get" function we extract the Sure name
tit["Name"].str.split(",").str.get(0)
```

#### <font color="red">REAL TIME EXAMPLE-2
- We want the Title & Full Name from "Name" Column

```python
# It give Full Name with Title
tit["Name"].str.split(",").str.get(1)
```

```python
# Every Name have Space from the First.
# So we First remove this Spaces.
tit["Name"].str.split(",").str.get(1).str.strip()
```

```python
# Here the Problem is From the Second Elements of list, there are more than one Elements.
tit["Name"].str.split(",").str.get(1).str.strip().str.split(" ")
```

```python
# It Split only First Occurance.
tit["Name"].str.split(",").str.get(1).str.strip().str.split(" ",n=1)
```

```python
# It extract the Title & Full Name and Store it to a dataframe
tit["Name"].str.split(",").str.get(1).str.strip().str.split(" ",n=1,expand=True)
```

```python
# Store it to the datasets
tit[["Title","Full Name"]]=tit["Name"].str.split(",").str.get(1).str.strip().str.split(" ",n=1,expand=True)
tit
```

#### <font color="Purple">FILTERING

#### <font color="Green"> 8. STRATSWITH
- It give the value that start with your Custom Value.
- eg:<br>
    `["cat","mat","rat","com"] ---> "cat" , "com"` <br>
    Here we want the elements that start with `"c"`.
    

```python
# It give all the Full Name that Starts with "A"
tit[tit["Full Name"].str.startswith("A")]
```

#### <font color="Green"> 9. ENDSWITH
- It give the value that start with your Custom Value.
- eg:<br>
    `["cat","mat","rat","com"] ---> "cat" , "mat" , "rat"` <br>
    Here we want the elements that end with `"t"`.
    

```python
# It give all the Full Name that Ends with "A"
tit[tit["Full Name"].str.endswith("A")]
```

#### <font color="green">10. ISDIGIT
- It give the values that have digits.
- eg:
    `E102`

```python
# It give the Ticket Column having Digit only.
tit[tit["Ticket"].str.isdigit()]
```

#### <font color="green">10. ISALPHA
- It give the values that have alphabets.
- eg:
    `ER202`

```python
# It give the Ticket Column having Alphabets only.
tit[tit["Ticket"].str.isalpha()]
```

#### <font color="green">11. CONTAINS
- `"CASE"--> Default is `__True__`. If you convert to `__False__` then it do not care about the CASE OF ALPHABETS.`
- It find the Particular Values that you provide.
- eg: <br>
    `["Rama","Samya","Ram","Dhama"] --> "Ram"` <br>
    Here we want to find a particular element i.e. `"Ram"` 

```python
# It give all the Full Name that Ends with "A"
tit[tit["Full Name"].str.contains("john",case=False)]
```

```python
# Find lastnames with start and end char vowel
tit[tit['Full Name'].str.contains('^[^aeiouAEIOU].+[^aeiouAEIOU]$')]
```
