### Merge

```python
dict8={
    "City":["Simla","Delhi","Jaypur","Bhopal"],
    "Temp":[25,56,34,23]
}

dict9={
    "City":["Chennai","Hyderabad","Vizag","Kochi"],
    "Humid":[67,75,83,54]
}
```

```python
d1=pd.DataFrame(dict8)
d2=pd.DataFrame(dict9)
```

```python
display(d1,d2)
```

```python
dict5={
    "City":["Chennai","Hyderabad","Vizag","Kochi"],
    "Temp":[25,56,34,23]
}

dict6={
    "City":["Chennai","Hyderabad","Vizag","Kochi"],
    "Humid":[67,75,83,54]
}
```

```python
d3=pd.DataFrame(dict5)
d4=pd.DataFrame(dict6)
```

```python
display(d3,d4)
```

```python
pd.merge(d3,d4, on="City")
```

```python
dict1={
    "City":["Chennai","Hyderabad","Vizag","Kochi"],
    "State":["TN","AP","AS","TS"],
    "Temp":[25,56,34,23]
}

dict2={
    "City":["Chennai","Hyderabad","Vizag","Kochi"],
    "State":["TN","AP","AS","TS"],
    "Humid":[67,75,83,54]
}
```

```python
d1=pd.DataFrame(dict1)
d2=pd.DataFrame(dict2)
```

```python
display(d1,d2)
```

```python
pd.merge(d1,d2)
```

```python
pd.merge(d1,d2, on="City")
```

```python
pd.merge(d1,d2, on="City",suffixes=("_left","_right"))
```

```python
dict3={
    "City":["Chennai","Hyderabad","Vizag","Kochi","Mumbai"],
    "Temp":[25,56,34,23,34]
}

dict4={
    "City":["Chennai","Hyderabad","Vizag","Kochi","Vopal"],
    "Humid":[67,75,83,54,46]
}
```

```python
d3=pd.DataFrame(dict3)
d4=pd.DataFrame(dict4)
```

```python
display(d3,d4)
```

```python
pd.merge(d3,d4) # Inner Join
```

```python
pd.merge(d3,d4,how="inner") # Inner Join
```

```python
pd.merge(d3,d4, how="left") # Left Centric Outer Join
```

```python
pd.merge(d3,d4, how="right") # Right Centric Outer Join
```

```python
pd.merge(d3,d4, how="cross") # Cross Join
```

```python
dict3={
    "City":["Chennai","Hyderabad","Vizag","Kochi"],
    "Temp":[25,56,34,23]
}

dict4={
    "City":["Bhopal","Mumbai","Bangalore","Kolkata"],
    "Humid":[67,75,83,54]
}

d11=pd.DataFrame(dict3)
d22=pd.DataFrame(dict4)
```

```python
display(d11,d22)
```

```python
pd.merge(d11,d22)
```

```python
pd.merge(d11,d22, how="outer") # Outer Join
```

### Join

```python
dict8={
    "NCity":["Simla","Delhi","Jaypur","Bhopal"],
    "Temp":[25,56,34,23]
}

dict9={
    "SCity":["Chennai","Hyderabad","Vizag","Kochi"],
    "Humid":[67,75,83,54]
}

d1=pd.DataFrame(dict8)
d2=pd.DataFrame(dict9)
```

```python
display(d1,d2)
```

```python
d1.join(d2)
```

### Append

```python
dict10={
    "City":["Simla","Delhi","Jaypur","Bhopal"],
    "Temp":[25,56,34,23]
}

dict11={
    "City":["Chennai","Hyderabad","Vizag","Kochi"],
    "Temp":[67,75,83,54]
}

d1=pd.DataFrame(dict10)
d2=pd.DataFrame(dict11)
```

```python
display(d1,d2)
```

```python
pd.concat([d1,d2])
```

```python
d1.append(d2)
```
