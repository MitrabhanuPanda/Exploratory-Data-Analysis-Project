### <font color="brown">STEP-3: 
        CATEGORIZE THEN INTO CAT-A,CAT-B AND CAT-C
        
    
        GET COUNTRY LEVEL AVG SALES--->
    
        IF THE TOTAL_UNITS_SOLD IS >= 1.25 OF AVG_UNITS_SOLD	CAT-A
		IF THE TOTAL_UNITS_SOLD IS >= .50 OF AVG_UNITS_SOLD	CAT-B
		IF THE TOTAL_UNITS_SOLD IS < .50 OF AVG_UNITS_SOLD	CAT-C


```python
cpts
```

```python
cpts["category"]="CAT-D"
cpts["category"][cpts["total_units_sold"]<( 0.50 * cpts["avg_units_sold"])]="CAT-C"
cpts["category"][cpts["total_units_sold"]>=( 0.50 * cpts["avg_units_sold"])]="CAT-B"
cpts["category"][cpts["total_units_sold"]>=( 1.25 * cpts["avg_units_sold"])]="CAT-A"
```
