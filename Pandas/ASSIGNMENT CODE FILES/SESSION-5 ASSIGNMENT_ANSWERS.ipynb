```python
df_task_3_summ=df_task_3.groupby(["COUNTRY","CATEGORY","PRODUCT_NAME"],as_index=False).agg(Total_Sales=("SALES","sum"))
```

```python
df_task_3_summ["Cummilative_sum"]=df_task_3_summ.groupby(["COUNTRY","CATEGORY"])["Total_Sales"].apply(lambda x:x.cumsum())
```

```python
df_task_3_summ
```

```python
df_task_3_summ2=df_task_3_summ.groupby(["COUNTRY","CATEGORY"],as_index=False).agg(total_sales_cat=("Total_Sales","sum"))
```

```python
df_task_3_summ2
```

```python
df_task_3_summ_join=pd.merge(left=df_task_3_summ,right=df_task_3_summ2,on=["COUNTRY","CATEGORY"],how="left")
```

```python
df_task_3_summ_join
```

```python
df_task_3_summ_join.columns
```

```python
df_task_3_summ_join["% of Category"]=np.round((df_task_3_summ_join["Total_Sales"]/df_task_3_summ_join["total_sales_cat"])*100,2)
```

```python
df_task_3_summ_join
```

```python
df_task_3_summ3=df_task_3_summ.groupby(["COUNTRY"],as_index=False).agg(total_sales_coun=("Total_Sales","sum"))
```

```python
df_task_3_summ3
```

```python
df_task_3_summ_join_1=pd.merge(left=df_task_3_summ_join,right=df_task_3_summ3,on=["COUNTRY"],how="left")
```

```python
df_task_3_summ_join_1
```

```python
df_task_3_summ_join_1.columns
```

```python
df_task_3_summ_join_1["% of Country"]=np.round((df_task_3_summ_join_1["Total_Sales"]/df_task_3_summ_join_1["total_sales_coun"])*100,2)
```

```python
df_task_3_summ_join_1
```

```python
df_task_3_summ_join_1.columns
```

```python
final_res=df_task_3_summ_join_1[["COUNTRY","CATEGORY","PRODUCT_NAME","Total_Sales","Ranking","Cummilative_sum","% of Category","% of Country"]]
```

```python
final_res
```
