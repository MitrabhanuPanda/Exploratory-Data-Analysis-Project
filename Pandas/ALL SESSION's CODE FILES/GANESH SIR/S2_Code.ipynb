# HOW TO FILTER DATA FROM A DATASET USING CONDITIONS

```python
med_2022
```

```python
x=['CUSTOMER_ID','Company','GENDER','Age','STATE_CODE','TOWN','NO_OF_TRIPS']
med_select=med_2022[x][(med_2022["Company"]=="APPOLO")]
med_select
```

```python
x=['CUSTOMER_ID','Company','GENDER','Age','STATE_CODE','TOWN','NO_OF_TRIPS']
med_select=med_2022[x][(med_2022["Company"]=="APPOLO") | (med_2022['GENDER']=='Male')]
med_select
```

```python
#med_2022['Company']=="APPOLO"
#Company is an operand
# == is an operator
#'APPOLO' is the conditional value
```

```python
#There are 4 types of operators
#1.LOGICAL OPERATOR (AND, OR , IN, NOT IN)
#2-COMPARISION OPERATOR (>,<.<=, <>)
#3- SPECIAL OPERATOR (LIKE, BETWEEN, AND)
#4- ARITHMATIC OPERATOR (+,-,*,/)
```

```python
#1.LOGICAL OPERATOR (AND, OR , IN, NOT IN)
```

```python
x=['CUSTOMER_ID','Company','GENDER','Age','STATE_CODE','TOWN','NO_OF_TRIPS']
med_select=med_2022[x][(med_2022["Company"]=="APPOLO") & (med_2022['GENDER']=='Male')]
med_select
```

```python
#OR OPERATOR
x=['CUSTOMER_ID','Company','GENDER','Age','STATE_CODE','TOWN','NO_OF_TRIPS']
med_select=med_2022[x][(med_2022["Company"]=="APPOLO") | (med_2022['GENDER']=='Male')]
med_select
```

```python
# IN OPERATOR
x=['CUSTOMER_ID','Company','GENDER','Age','STATE_CODE','TOWN','NO_OF_TRIPS']
med_select=med_2022[x][(med_2022["Company"].isin(['APPOLO','CIPLA','GSK','GENO']))]
med_select
```

```python
#not in operator
x=['CUSTOMER_ID','Company','GENDER','Age','STATE_CODE','TOWN','NO_OF_TRIPS']
med_select=med_2022[x][(med_2022["Company"].isin(['APPOLO','CIPLA','GSK','GENO']))==False]
med_select
```

```python
#2-COMPARISION OPERATOR (>,<.<=, <>)
```

```python
x=['CUSTOMER_ID','Company','GENDER','Age','STATE_CODE','TOWN','NO_OF_TRIPS']
med_select=med_2022[x][(med_2022["Company"]=="APPOLO") & (med_2022['GENDER']=='Male') & (med_2022['Age'] > 40)]
med_select
```

```python
x=['CUSTOMER_ID','Company','GENDER','Age','STATE_CODE','TOWN','NO_OF_TRIPS']
med_select=med_2022[x][(med_2022["Company"]=="APPOLO") & (med_2022['GENDER']=='Male') & (med_2022['Age'] <= 60)]
med_select
```

```python
#3- SPECIAL OPERATOR (LIKE, BETWEEN)
x=['CUSTOMER_ID','Company','GENDER','Age','STATE_CODE','TOWN','NO_OF_TRIPS']
med_select=med_2022[x][med_2022['TOWN'].str.startswith('C',na=False)]
med_select
```

# SUMMARIZING DATASET BY DIMENSION AND THE MEASURE FIELDS

```python
#to get company wise total number of visits
med_2022['NO_OF_TRIPS'].groupby(med_2022['Company']).sum()
```

```python
#to get company wise avg age of customers
med_2022['Age'].groupby(med_2022['Company']).mean()
```

```python
# get state_code and company wise customer base
med_2022['CUSTOMER_ID'].groupby([med_2022['STATE_CODE'],med_2022['Company']]).count()
```

```python
med_summary=med_2022.groupby(['STATE_CODE','Company'], as_index=False).agg(SUBS=('CUSTOMER_ID','count'))
med_summary
```

```python
med_summary1=med_2022.groupby(['STATE_CODE','Company'], as_index=True).agg(SUBS=('CUSTOMER_ID','count'))
med_summary1
```

```python
#to get state_code, company wise total_customers, total_visits
med_summary=med_2022.groupby(['STATE_CODE','Company'], as_index=False).agg(SUBS=('CUSTOMER_ID','count'),Total_visit=('NO_OF_TRIPS','sum'))
med_summary
```
