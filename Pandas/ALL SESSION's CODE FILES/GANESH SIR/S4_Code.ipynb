# Ranking, cummulative sum, looping, first() and last() function

```python
#GET the state_code wise total customer base

med_summary=med_2023.groupby(['STATE_CODE'],as_index=False).agg(Customer_Base=('CUSTOMER_ID','count'))
med_summary
```

```python
#Rank STATE_CODE wise customer base

med_summary['rankings']=med_summary['Customer_Base'].rank()
med_summary
```

```python
med_summary['rankings1']=med_summary['Customer_Base'].rank(method='dense',ascending=False).astype(int)
med_summary
```

```python
#Ranking the records based on two columns

#Get the ranks to the customer base based on state_code and company

med_summary=med_2023.groupby(['STATE_CODE','Company'],as_index=False).agg(Customer_Base=('CUSTOMER_ID','count'))
med_summary





```

```python
med_summary['rankings']=med_summary.groupby('STATE_CODE')['Customer_Base'].rank(method='dense',ascending=True).astype(int)
med_summary
```

# Cummulative sum calculation

```python
med_summary=med_2023.groupby(['STATE_CODE','Company'],as_index=False).agg(Customer_Base=('CUSTOMER_ID','count'))
med_summary
```

```python
#Get the cummulative summ by 2 dimensional field

med_summary['cumulative_base']=med_summary.groupby(['STATE_CODE'])['Customer_Base'].apply(lambda x: x.cumsum())
med_summary[0:10]
```

# NUMERIC FUNCTIONS

```python
import pandas as pd
prod_sales=pd.DataFrame({'product':['apple','dell','lenovo','hp','acer','samsung'],
                     'jan_sales':[200,200,500,900,230,560],
                     'feb_sales':[200,340,500,450,230,900],
                     'mar_sales':[240,200,500,980,230,560],
                     'apr_sales':[200,200,490,900,230,510],
                     'may_sales':[300,200,500,900,730,560],
                     'jun_sales':[600,200,500,500,230,510]})
prod_sales
```

```python
prod_sales_v1=prod_sales.copy(deep=True)
prod_sales_v1
```

```python
#How to aggregate the values
#Sum
#AVG
#MIN
#MAX
```

```python
prod_sales_v1['total_sales']=prod_sales_v1[['jan_sales','feb_sales','mar_sales','apr_sales','may_sales','jun_sales']].sum(axis=1)
```

```python
prod_sales_v1
```

```python
prod_sales_v1['avg_sales']=prod_sales_v1[['jan_sales','feb_sales','mar_sales','apr_sales','may_sales','jun_sales']].mean(axis=1)
prod_sales_v1['min_sales']=prod_sales_v1[['jan_sales','feb_sales','mar_sales','apr_sales','may_sales','jun_sales']].min(axis=1)
prod_sales_v1['max_sales']=prod_sales_v1[['jan_sales','feb_sales','mar_sales','apr_sales','may_sales','jun_sales']].max(axis=1)
prod_sales_v1
```
