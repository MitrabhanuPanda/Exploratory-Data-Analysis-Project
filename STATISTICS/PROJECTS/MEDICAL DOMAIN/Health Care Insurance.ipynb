### <font color='green'> OUTLIER DETECTION

#### MEAN-STD METHOD :
#### LOWER LIMIT= MEAN-(3*STD)
#### UPPER LIMIT= MEAN+(3*STD)

```python
mean_bmi=df['bmi'].mean() # Mean value of BMI feature
```

```python
std_bmi=df['bmi'].std() # Standard Deviation of BMI feature
```

```python
lower_limit=mean_bmi-(3*std_bmi) # Lower Limit of BMI feature
```

```python
upper_limit=mean_bmi+(3*std_bmi) # Upper Limit of BMI feature
```

```python
df[df['bmi']<lower_limit]
```

```python
df[df['bmi']>upper_limit]
```

```python
df[(df['bmi']<lower_limit)|(df['bmi']>upper_limit)] # Show records of the outliers
```

### Trimming

```python
df=df[((df['bmi']>lower_limit))&((df['bmi']<upper_limit))] # Delete records of the outliers and put into the dataset
```

### <font color='green'> OUTLIER DETECTION

#### q1=df[column name].quantile(0.25)
#### q3=df[column name].quantile(0.75)

#### IQR= q3-q1

#### upper limt=q1-(1.5*IQR)
#### lower limt=q3+(1.5*IQR)

```python
q1=df['charges'].quantile(0.25)
q3=df['charges'].quantile(0.75)

IQR= q3-q1

charges_lower=q1-1.5*IQR
charges_upper=q3+1.5*IQR
```

```python
df[(df['charges'] < charges_lower) | (df['charges'] > charges_upper)]
```

```python
df['charges'] = df['charges'].map( lambda x : charges_lower  if x < charges_lower else  charges_upper if x > charges_upper else x)
```

```python
df[(df['charges'] < charges_lower) | (df['charges'] > charges_upper)]
```
