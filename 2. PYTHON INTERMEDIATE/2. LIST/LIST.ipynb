<h1 style="text-align : center"> <font color="red" size=8>LIST </h1>

- LIST IS A DATA TYPE WHERE YOU CAN STORE MULTIPLE ITEMS UNDER 1 NAME.
- MORE TECHNICALLY, LISTS ACTS AS DYNAMIC ARRAY WHICH MEANS YOU CAN ADD MORE ITEMS ON THE FLY.

### <font color="brown">WHY LISTS ARE REQUIRED IN PROGRAMMING? <font color/>
- It is sequential data type which holds the multiple heterogeneous value.
- These value are closed by square bracket[] separated by comma.
- These value inside the square brackets are reffered as __`Elements`__
- Using List function, we can convert any value into list.

### <font color="brown">FEATURES:<font color/>
- Multiple Heterogeneous Values
- Mutable(Changeable)
- Duplicate are allows
- Order is preserved(Input Order is preserved)
- Indexing is possible
- Slicing is possible
- Nested list is possible

## <font color="teal">REFERENCE COPY:<font color/>
* It is a process of creating an instance of the original data in a different variable. 
* But in return both variables are pointing to same memory location.
* Assigning a value within another variable.

```python
a1=[10,20,30,40,50]
print(a1)
print(id(a1))
```

```python
# Reference Copy

b1=a1

print(b1)
print(id(b1))
```

```python
b1.append("SKY")

print(b1)
```

```python
print(a1)
```

```python
a1.append("BIRD")

print(a1)
```

```python
print(b1)
```

# <font color="dark blue" font size=7> ZIP FUNCTION<font color/>

- The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together.

- If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator

# <font color="brown">HOW TO ADD TWO LIST'S ITEMS? <font color/>

```python
u1=[1,2,3,4,5]
u2=[6,7,8,9,0]
u3=[8,9,0]
```

### <font color="dark red">TWO LIST HAVE SAME SIZE<font color="dark red">

```python
for i,j in zip(u1,u2):
    print(i,j)
```

```python
for i,j in zip(u1,u2):
    print(i+j)
```

```python
res=[]
for i,j in zip(u1,u2):
    res.append(i+j)
```

```python
res
```

### <font color="dark red">TWO LIST HAVE DIFFERENT SIZE<font color="dark red">

```python
# IF TWO LISTS HAVING DIFFERENT SIZE, IT FIRST SEE WHAT IS SIZE LESS SIZE, THEN ADD ITEMS WITH THAT MUCH SIZE.
for i,j in zip(u1,u3):
    print(i,j)
```

```python
res1=[]
for i,j in zip(u1,u3):
    res1.append(i+j)
```

```python
res1
```

# <font color="dark blue" font size=10> LIST COMPREHENSION<font color/>

- LIST COMPREHENSION PROVIDES A CONCISE WAY OF CREATING LISTS.

## <font color="blue">ADVANTAGES OF LIST COMPREHENSION<font color/>
- MORE TIME-EFFICIENT & SPACE-EFFICIENT THAT LOOPS.
- REQUIRE FEWER LINES OF CODE.
- TRANSFORMS ITERATIVE STATEMENT INTO A FORMULA.

```python
[x for x in range(1,6)] # 1-5
```

```python
[i for i in range(1,10)]
```

```python
[x**3 for x in range(1,6)]
```

```python
[x*x for x in range(1,11)] # 1-10 square values in list
```

```python
n=int(input("Enyer a Number: "))
[i*n for i in range(1,11)]
```

```python
[str(n)+ "*"+ str(i)+ "=="+ str(i*n) for i in range(1,11)]
```

## <font color="dark red">LIST COMPREHENSION WITH MULTIPLE FOR LOOP<font color="dark red">

```python
rnum=int(input("Enter a Number: "))

[str(x)+"*"+str(i)+"=="+str(x*i)   for x in range(2,rnum+1) for i in range(1,11)]
```

## <font color="dark red">LIST COMPREHENSION WITH IF CONDITION<font color="dark red">

```python
n1=int(input("Enter a Number: "))
[i for i in range(1,n1+1) if i%2==0]
```

```python
n1=int(input("Enter a Number: "))
[i*i for i in range(1,n1+1) if i%2==0]
```

### <font color="orange">LIST COMPREHENSION WITH MULTIPLE IF CONDITION<font color="dark red">

```python
n1=int(input("Enter a Number: "))
[i for i in range(1,n1+1) if i%3==0  if i%5==0] 
```

## <font color="dark red">LIST COMPREHENSION WITH IF..ELSE CONDITION<font color="dark red">

```python
n1=int(input("Enter a Number: "))
[str(i) + " is Even"  if i%2==0   else str(i) + " is Odd"       for i in range(1,n1+1) ] 
```

## <font color="dark red">NESTED LIST COMPREHENSION<font color="dark red">

## <font color="dark red">EXAMPLES LIST COMPREHENSION<font color="dark red">

```python
# find languages which start with letter p
languages = ['java','python','php','c','javascript']

[language for language in languages if language.startswith('p')]
```

```python
# Nested if with List Comprehension
basket = ['apple','guava','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']

# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'

[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

```python
# scalar multiplication on a vector
v = [2,3,4]
s = -3
# [-6,-9,-12]

[s*i for i in v]
```
