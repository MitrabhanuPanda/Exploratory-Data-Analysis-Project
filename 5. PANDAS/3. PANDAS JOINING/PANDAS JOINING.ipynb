## <font color="brown">JOIN--> MERGE
    If you have one column common in both Datasets

#### INNER JOIN

```python
pd.merge(std,nov,on="student_id",how="inner")
```

#### LEFT JOIN

```python
cor.merge(nov,how="left",on="course_id")
```

#### LEFT NULL JOIN

```python
dfff=cor.merge(nov,how="left",on="course_id")
lef_nul_df=dfff[dfff["student_id"].isnull()]
lef_nul_df
```

#### RIGHT JOIN

```python
df_rgh=cor.merge(nov,how="right",on="course_id")
df_rgh
```

#### RIGHT NULL JOIN

```python
righ_nul_df=df_rgh[df_rgh["course_name"].isnull()]
righ_nul_df
```

#### FULL OUTER JOIN

```python
ful_df=cor.merge(nov,how="outer",on="course_id")
ful_df
```

#### OUTER JOIN

```python
out_df=ful_df[(ful_df["course_name"].isnull()) | (ful_df["student_id"].isnull())]
out_df
```
