## <font color="dark blue">PIVOT TABLE
- THE PIVOT TABLE TAKES SIMPLE COLUMN-WISE DATA AS INPUT & GROUPS THE ENTIRES INTO A TWO-DIMENSIONAL TABLE THAT PROVIDES A MULTIDIMENSIONAL SUMMARIZATION OF THE DATA.

- INDEX--> TO WHICH VARIABLE YOU WANT TO DO YOUR ROW/INDEX. (CATEGORICAL/DIMENSION FIELD)

- COLUMNS---> TO WHICH VARIABLE YOU WANT TO DO YOUR COLUMN. (CATEGORICAL/DIMENSION FIELD)

- VALUES---> TO WHICH VARIABLE YOU WANT TO SHOW AS VALUES. (MEASURE/NUMERIC FIELD)

- AGGFUN---> BY DEFAULT "MEAN" YOU CAN CONVERT TO (SUM,STD,COUNT)

- MARGIN--> IN GIVE TOTAL/SUM OF VALUES ==> (True)

- DROPNA--> IT DELETE THE "NAN".

- FILL_VALUE---> IT FILL THE VALUE THAT YOU PROVIDE TO "NAN" PLACES.

- MARGIN_NAME--> BY DEFAULT "ALL". YOU CAN CHANGE IT.

- SORT--> BY DEFAULT "TRUE". IF YOU CHANGE IT TO "FALSE" THEN THE VALUES SORT INTO DESCENDING ORDER.

```python
# IT GIVE EVERY COLUMN WISE ANALYSIS
df.pivot_table(index="sex",columns="smoker")
```

```python
# SEX & SMOKER WISE AVERGAE TOTAL_BIL
df.pivot_table(index="sex",columns="smoker",values="total_bill")
```

```python
# SEX & SMOKER WISE TOTAL_BIL
df.pivot_table(index="sex",columns="smoker",values="total_bill",aggfunc="sum")
```

```python
# SEX,SMOKER & DAY,TIME WISE AVERAGE TOTAL BIIL
df.pivot_table(index=["sex","smoker"],columns=["day","time"],values="total_bill")
```

```python
# SEX,SMOKER & DAY,TIME WISE AVERAGE MAX BILL & AVERAGE SIZE & TOTAL TIP
df.pivot_table(index=["sex","smoker"],columns=["day","time"],aggfunc={"size":"mean","tip":"sum","total_bill":"max"})
```

```python
# SEX & SMKOER WISE TOTAL BILL WITH TOTAL OF VALUES
df.pivot_table(index="sex",columns="smoker",values="total_bill",aggfunc="sum",margins=True)
```

```python
# NAMING THE MARGIN
df.pivot_table(index="sex",columns="smoker",values="total_bill",aggfunc="sum",margins=True,margins_name="Total")
```

```python
# SORT IT
df.pivot_table(index="sex",columns="smoker",values="total_bill",aggfunc="sum",margins=True,margins_name="Total",sort=False)
```
