### <font color="orange">AGGREGARTION METHOD USES

#### PASSING IN DICTIONARY

```python
gen.agg({
    'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
})
```

#### PASSING LIST

```python
gen.agg(["min","max","sum","mean"])
```

#### PASSING IN BOTH DICTIONARY & LIST

```python
gen.agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```
