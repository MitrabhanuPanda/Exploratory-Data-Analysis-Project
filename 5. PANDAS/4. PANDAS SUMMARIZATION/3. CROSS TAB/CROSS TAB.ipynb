```python
# Index--> Row 
# Columns --> Column

pd.crosstab(index=df["NATIONALITY"],columns=df["HANDEDNESS"])
```

```python
# Margins --> It give the Grand Total For both Columns & Rows

pd.crosstab(index=df["NATIONALITY"],columns=df["HANDEDNESS"],margins=True)
```

```python
# normalize --> It is used to give the Numeric Values in Percentage Format
pd.crosstab(index=df["NATIONALITY"],columns=df["HANDEDNESS"],margins=True,normalize=True)
```

```python
# Percentage Values in Index Level
pd.crosstab(index=df["NATIONALITY"],columns=df["HANDEDNESS"],margins=True,normalize="index")
```

```python
# Percentage Values in Column Level
pd.crosstab(index=df["NATIONALITY"],columns=df["HANDEDNESS"],margins=True,normalize="columns")
```

```python
# By default, Cross tab give the Frequency.
# If you want to show the values in other Mathematical form like Sum, Mean, Median, Standard Deviation, then use "aggfunc".
# You can not use "aggfun" without "values".

# In "values", you pass the 'Numeric Features' only.
pd.crosstab(index=df["NATIONALITY"],columns=df["HANDEDNESS"],margins=True,values=df["AGE"],aggfunc="mean")
```

```python
pd.crosstab(index=df["NATIONALITY"],columns=df["HANDEDNESS"],margins=True,values=df["AGE"],aggfunc="median")
```

```python
pd.crosstab(index=df["NATIONALITY"],columns=df["HANDEDNESS"],margins=True,values=df["AGE"],aggfunc="sum")
```

```python
pd.crosstab(index=df["NATIONALITY"],columns=df["HANDEDNESS"],margins=True,values=df["AGE"],aggfunc="std")
```

```python
# You can also give Custom name to the Margins using "margins_name"
pd.crosstab(index=df["NATIONALITY"],columns=df["HANDEDNESS"],margins=True,values=df["AGE"],aggfunc="std",margins_name="Grand Total")
```

```python
# We can also pass more than one aggregate functions (mean,std) at a time in a 'list'.
pd.crosstab(index=df["NATIONALITY"],columns=df["HANDEDNESS"],margins=True,values=df["AGE"],aggfunc=["std","mean"],margins_name="Grand Total")
```

```python
# We can also give more than one categorical features in "columns".
# For this,We have to write the features name inside a list.

pd.crosstab(index=df["NATIONALITY"],columns=[df["HANDEDNESS"],df["SEX"]],margins=True)
```

```python
# We can also give more than one categorical features in "index".
# For this,We have to write the features name inside a list.

pd.crosstab(index=[df["NATIONALITY"],df["SEX"]],columns=[df["HANDEDNESS"]],margins=True)
```
