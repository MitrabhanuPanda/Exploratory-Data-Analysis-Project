## <font color="brown">ANNOTATION</font>

#### `plt.text(x1,y1,name)`

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.scatter(x,y)
plt.text(1,5,'Point 1')
plt.text(2,6,'Point 2')
plt.text(3,7,'Point 3')
plt.text(4,8,'Point 4',fontdict={'size':12,'color':'brown'})
```

####  `plt.ahline(y_cordinate_value,color="red"/"greeen")`

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.scatter(x,y)

plt.axhline(6.5,color='red')
plt.axhline(7.0,color='green')
plt.axvline(2.5,color='red')

plt.text(1,5,'Point 1')
plt.text(2,6,'Point 2')
plt.text(3,7,'Point 3')
plt.text(4,8,'Point 4',fontdict={'size':12,'color':'brown'})
```

## <font color="brown">SUB-PLOTS</font>
* It is nothing but another way to plot graph.
* You can also plot many graph using SubPlot.

```python
plt.scatter(sample_df["avg"],sample_df["strike_rate"])
```

```python
plt.subplots()

 # It give two objects. (Figure & Axes)
```

```python
fig,ax = plt.subplots()  

ax.scatter(sample_df["avg"],sample_df["strike_rate"])   # Plot graph using axes object
ax.set_title("Batter")    # Add Title
ax.set_xlabel("Avg")    # Add x-axis lable
ax.set_ylabel("Strike_rate")   # Add y-axis lable

fig.show() 
```

```python
fig,ax = plt.subplots(figsize=(15,6))  # Add width & height / Increase the size of the Graph

ax.scatter(sample_df["avg"],sample_df["strike_rate"])   # Plot graph using axes object
ax.set_title("Batter")    # Add Title
ax.set_xlabel("Avg")    # Add x-axis lable
ax.set_ylabel("Strike_rate")   # Add y-axis lable

```

### <font color="dark blue">HOW TO PLOT MULTIPLE GRAPHS</font>

```python
plt.subplots(nrows=2,ncols=1)   # Add 2 Blank Figures 
```

```python
plt.subplots(nrows=2,ncols=2)  # Add 4 Blank Figures 
```

### <font color="green">PLOTTING 2 GRAPHS</font>

```python
fig,ax = plt.subplots(nrows=2,ncols=1,figsize=(10,6))

ax[0].scatter(sample_df["avg"],sample_df["strike_rate"])
ax[1].scatter(sample_df["runs"],sample_df["strike_rate"])


ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')


ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1,sharex=True)

ax[0].scatter(sample_df["avg"],sample_df["strike_rate"])
ax[1].scatter(sample_df["runs"],sample_df["strike_rate"])
```

### <font color="green">PLOTTING 4 GRAPHS</font>

```python
fig,ax=plt.subplots(nrows=2, ncols=2)

ax[0,0].scatter(sample_df["avg"],sample_df["strike_rate"])
ax[0,1].scatter(sample_df["runs"],sample_df["strike_rate"])

ax[1,0].hist(sample_df["avg"])
ax[1,1].hist(sample_df["runs"])
```

### <font color="dark blue">ANOTHER WAY TO PLOT MULTIPLE GRAPHS</font>

#### `fig.add_subplot(number of rows,number of columns,Figure number)`

```python
fig=plt.figure()

ax1=fig.add_subplot(2,1,1)
ax1.scatter(sample_df["avg"],sample_df["strike_rate"])

ax2=fig.add_subplot(2,1,2)
ax2.scatter(sample_df["runs"],sample_df["strike_rate"])
```

```python
fig=plt.figure()

ax1=fig.add_subplot(2,2,1)
ax1.scatter(sample_df["avg"],sample_df["strike_rate"])

ax2=fig.add_subplot(2,2,2)
ax2.scatter(sample_df["runs"],sample_df["strike_rate"])

ax3=fig.add_subplot(2,2,3)
ax3.hist(sample_df["runs"])


ax4=fig.add_subplot(2,2,4)
ax4.hist(sample_df["avg"])
```
