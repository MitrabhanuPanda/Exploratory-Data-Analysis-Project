*  <font color="red">Problem:
>    - Some time we have big big name Categorical value at that time if we will plot the graph then the name of category overlap each other.
    
* <font color="blue">Solution:
>    - Use xticks() function.
>    - Syntax ==> xticks(rotation = 45 / "vertical")

```python
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children)
plt.xticks(rotation='vertical')
plt.show()
```

```python
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation=45)
plt.show()
```

Use of __`autopct`__ ==> <br>

> - It show the __`percentage`__ in the Pie chart. <br>
> - Syntax ==> autopct="%0.2F%%"

```python
plt.pie(gyl["batsman_runs"],labels=gyl["batsman"],autopct="%0.2F%%")
plt.show()
```

Use of __`explode`__  Parameter ==> <br>

> - You bring a particular section of pie chart to outside by using explode `parameter`.
> - You pass a list with decimal numbers. 
> - Eg: 
    >> [0,0.05,0.02,0] <br>
    >> 0 --> Not Explode <br>
    >> 0.05 --> Explode <br>
    >> 0.02 --> Explode <br>

```python
e = [0.05, 0.03, 0, 0, 0,0]
```

```python
plt.pie(gyl["batsman_runs"],labels=gyl["batsman"],autopct="%0.2F%%",explode=e)
plt.show()
```

#### <font color="orange">Coloring the Pie Chart

```python
c = ['r','y','g','b','m','r']
```

```python
plt.pie(gyl["batsman_runs"],labels=gyl["batsman"],autopct="%0.2F%%",colors=c)
plt.show()
```

Use of __`shadow`__ ==> <br>

- It reflect the Shadow of the Pie Chart
- Syntax --> shadow = `True`

```python
plt.pie(gyl["batsman_runs"],labels=gyl["batsman"],autopct="%0.2F%%",explode=e,colors=c,shadow=True)
plt.show()
```

Use of __`radius`__ ==> <br>

    It increase the size of the Pie Chart
    Syntax --> radius = 1.4 / 2.5

```python
plt.pie(gyl["batsman_runs"],labels=gyl["batsman"],autopct="%0.2F%%",explode=e,colors=c,shadow=True,radius = 1.4)
plt.show()
```
