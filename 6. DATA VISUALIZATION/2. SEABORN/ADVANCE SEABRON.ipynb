## <font color="brown">1. CATEGORICAL SCATTER PLOT</font>
* USED FOR BI-VARIATE ANALYISIS.
* ONE CATEGORICAL & ONE NUMERICAL COLUMN IS REQUIRED.

### <font color="dark blue">a. STRIP PLOT</font>

```python
# AXIS LEVEL FUNCTION
sns.stripplot(data=tips,x="day",y="total_bill")
```

```python
# JITTER PARAMETER
sns.stripplot(data=tips,x="day",y="total_bill",jitter=False)
```

```python
# HUE PARAMETER
sns.stripplot(data=tips,x="day",y="total_bill",hue="sex")
```

```python
# FIGURE LEVEL FUNCTION
sns.catplot(data=tips,x="day",y="total_bill",kind="strip")
```

```python
# JITTER
sns.catplot(data=tips,x="day",y="total_bill",kind="strip",jitter=0.2)
```

### <font color="dark blue">b. SWARM PLOT</font>

```python
# AXIS LEVEL FUNCTION
sns.swarmplot(data=tips,x="day",y="total_bill")
```

```python
# FIGURE LEVEL FUNCTION
sns.catplot(data=tips,x="day",y="total_bill",kind="swarm")
```

```python
sns.catplot(data=tips,x="day",y="total_bill",kind="swarm",hue="sex")
```

## <font color="brown">2. CATEGORICAL DISTRIBUTION PLOT</font>
* USED FOR UNIVARIATE ANALYSIS

### <font color="dark blue">a. BOX PLOT</font>
* A Box plot is a Standardized way of displaying the distribution of data based on a Five Number Summary (Minimum, First Quartile[Q1], Median, Third Quartile[Q3] & Maximum ).


* It can tell you about your Outliers & What their Values are.


* Box plots can also tell if your data is Symmetrical, How tightly your data is grouped & if & how your data is skewed.

```python
# ONE NUMERICAL PLOT & ONE CATEGORICAL PLOT
sns.boxplot(data=tips,x="sex",y="total_bill")
```

```python
sns.boxplot(data=tips,x="day",y="total_bill")
```

```python
# BOX PLOT USING CATPLOT
sns.catplot(data=tips,x="day",y="total_bill",kind="box")
```

```python
# HUE PARAMETER
sns.catplot(data=tips,x="day",y="total_bill",kind="box",hue="sex")
```

```python
# NUMERICAL COLUMN
sns.boxplot(data=tips,y="total_bill")
```

### <font color="dark blue">b. VIOLIN PLOT (BOX PLOT + KDE PLOT)</font>
* __KDE PLOT__ --> It tells the DIstribution of Column.


* __BOX PLOT__ --> It tells about the outliers, skewness of a Column.

```python
sns.violinplot(data=tips,x="day",y="total_bill")
```

```python
# USING CATPLOT
sns.catplot(data=tips,x="day",y="total_bill",kind="violin")
```

```python
# HUE PARAMETER
sns.catplot(data=tips,x="day",y="total_bill",kind="violin",hue="sex")
```

```python
# SPLIT PARAMETER
sns.catplot(data=tips,x="day",y="total_bill",kind="violin",hue="sex",split=True)
```

### 

## <font color="brown">3. CATEGORICAL ESTIMATE PLOT</font>

### <font color="dark blue">a. BAR PLOT</font>
* One Categorical & One Numerical Column is Required.
* It give the Average of the numerical Column.

```python
sns.barplot(data=tips,x="sex",y="total_bill")
```

#### <font color="orange">BLACK COLOR BAR IN THE BAR PLOT:</font>
* It is called as __`Error Bar`__.
* It is plot when your categorical values are repeated.
* When there are multiple observations in each category, Bar Plot also uses Bootstraping to compute a Confidence Interval around the Estimate, which is plotted using __`Error Bar`__.

```python
# IF YOU DO NOT WANT TO ADD ERROR BAR, THEN USE "ci" PARAMETER
sns.barplot(data=tips,x="sex",y="total_bill",ci=None)
```

```python
# HUE PARAMETER
sns.barplot(data=tips,x="sex",y="total_bill",hue="smoker")
```

```python
# USING CAT PLOT
sns.catplot(data=tips,x="sex",y="total_bill",kind="bar")
```

```python
# YOU CAN ALSO CHANGE THE ESTIMATOR OF MEAN TO OTHER MATHEMATICAL ESTIMATE
# USE "ESTIMATOR" PARAMETER

import numpy as np

sns.barplot(data=tips,x="sex",y="total_bill",hue="smoker",estimator=np.std)
```

```python
import numpy as np

sns.barplot(data=tips,x="sex",y="total_bill",hue="smoker",estimator=np.median)
```

### <font color="dark blue">b. POINT PLOT</font>
* It just Connect the Error Bar.

```python
sns.pointplot(data=tips,x="sex",y="total_bill",hue="smoker")
```

```python
sns.pointplot(data=tips,x="sex",y="total_bill",hue="smoker",ci=None)
```

### <font color="dark blue">c. COUNT PLOT</font>
* A special case for the bar plot is when you want to show the Number of Observations in each Category rather than computing a statistic for a second variable.
* This is similar to a histogram over a Categorical, rather than Quantitative variable.

* ONE CATEGORICAL VARIABLE IS REQUIRED

```python
sns.countplot(data=tips,x="sex")
```

```python
# HUE PARAMETER
sns.countplot(data=tips,x="sex",hue="day")
```

### 

### <font color="dark blue">REGRESSION PLOT</font>
   * REGPLOT --> AXIS LEVEL FUNCTION
   * LMPLOT --> FIGURE LEVEL FUNCTION
   
* In the simplest invocation, both functions draw a scatterplot of two variables, x & y, & then fit the Regression Model y~x & plot the resulting regression line & a 95% confidence interval for that regression.

```python
# AXIS LEVEL FUNCTION

# You can not add HUE parameter
sns.regplot(data=tips,x="total_bill",y="tip")
```

```python
# FIGURE LEVEL FUNCTION
sns.lmplot(data=tips,x="total_bill",y="tip")
```

```python
# HUE PARAMETER
sns.lmplot(data=tips,x="total_bill",y="tip",hue="sex")
```

### 

### <font color="dark blue">PAIR GRID</font>

```python
g=sns.PairGrid(data=iris)
# USES OF map
g.map(sns.scatterplot)
```

```python
g=sns.PairGrid(data=iris,hue="species")

g.map(sns.scatterplot)
```

```python
# USES OF map_diag & map_offdia

g=sns.PairGrid(data=iris,hue="species")

g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g=sns.PairGrid(data=iris,hue="species")

g.map_diag(sns.violinplot)
g.map_offdiag(sns.scatterplot)
```

```python
# USES OF map_lower & map_lower

g=sns.PairGrid(data=iris,hue="species")

g.map_diag(sns.violinplot)
g.map_lower(sns.kdeplot)
g.map_upper(sns.scatterplot)
```

```python
# USES OF vars

g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
