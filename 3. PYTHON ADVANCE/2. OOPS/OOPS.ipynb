## <font color="red">WHAT IS CONSTRUCTOR?</font>
* IT IS A SPECIAL METHOD.
* IT AUTOMATICALLY CALL WHEN THE OBJECT IS CREATED.
* A METHOD WHICH CAN BE EXECUTED WITHOUT EXPLICITLY CALLING.
* CONSTRUCTOR IS A SPECIAL TYPE OF METHOD WHICH IS EXECUTED WHENEVER AN OBJECT IS CREATED.
* THESE CONSTRUCTORS ARE USED TO INITIALIZE THE VARIABLES WITH VALUES.
* NAME OF CONSTRUCTOR IS : __`_ _ init _ _ ()`__
* CONSTRUCTOR COMES UNDER THE INSTANCE METHOD CATEGORY, i.e. DEFAULT ARGUMENT IS `self`.
* ANY NUMBER OF CONSTRUCTOR CAN BE DEFINED WITHIN A CLASS BUT ONLY ONE IS EXECUTED i.e. __`LATEST ONE`__.

     

## <font color="red">TRUE BENIFIT OF CONSTRUCTOR:</font>
* TO INITIALIZE THE OBJECT.
* TO WRITE CONFIGURATION RELATED CODE.
* THAT MEANS IF WE WOULD LIKE TO CONNECT TO OUR DATABASE, TO OUR BACKEND SERVICE, WE WRITE THIS CODE INSIDE THE CONSTRUCTOR.
* IN SIMPLE, WE DO NOT GIVE THE CODE CONTROL BY USER. IN A CLASS EVERY FUNCTION IS CONTROLLED BY THE USER BUT THE CONSTRUCTOR IS NOT CONTROLLED BY THE USER.


* eg:

    * __QUES__:
    
        IF GOD IS THE PROGRAMMER, EARTH IS THE CLASS, HUMANE BEINGS IS THE OBJECT THEN WHAT IS THE CONSTRUCTOR?
        
    * __ANS__:
    
       - THE CONSTRUCTOR IS DEATH.
       - GOD DO NOT WANT TO GIVE THE DEATH TO CONTROL BY THE HUMAN.

### <font color="brown">`"self"` KEYWOORD</font>
* __`self`__ REPRESENTS THE INSTANCE OF CLASS.


* IT ALLOWS US TO ACESS VARIABLES, ATTRIBUTES, & METHODS OF A DEFINED CLASS.


* IT REFERS TO CUREENT OBJECT.


* IT IS AN IDENTIFIER WHICH HOLDS OBJECT OF THE CLASS.


* IT HELPS TO IDENTIFY THE LOCAL & INSTANCE VARIABLES.

* IT GIVE THE POWER THAT IF WE WANT TO CALL A METHOD INSIDE ANOTHER METHOD IN THAT TIME WE USE __`self`__.


* __`self`__ IS NOTHING BUT THE CURRENT OBJECT.

```python
# eg

# Create a Class

class con_class:
    
    def __init__(self):   # Constructor --> Default
        print("This is from constructor method")
        self.n1()  
        
        
    def n1(self):  # Method
        print("This is from n1 method")
        num=5
        
        for i in range(1,11):
            print(num, "*", i, "==", (num*i))
```

### <font color="red">3rd MAGIC METHOD:</font>
   ` _ _ add _ _ ()` --> "+"
   
   - IT AUTOMATICALLY TRIGGER, WHEN YOU GIVE OPERAND "+" IN BETWEEN THE TWO OBJECTS OF THE CLASS.
   
   ` _ _ sub _ _ ()` --> "-"
   
   
   
   ` _ _ mul _ _ ()` --> "*"
   
   
   
   ` _ _ true div _ _ ()` --> "/"

```python
class fraction:
    
    def __init__(self,x,y):
        self.num=x
        self.den=y
        
    def __str__(self):
        print( "idhar")
        
    def __add__(self,other):
        
        new_num = (self.num * other.den) + (other.num * self.den)
        new_den = self.den + other.den
        
        return "{}/{}".format(new_num,new_den)
        
```

### <font color="blue">c. STATIC VARIABLES:</font>
* THESE VARIABLES ARE CONSIDERED AS A GLOBAL VARIABLE WHICH CAN BE ACCESSED OUTSIDE THE CLASS ALSO.
* THESE VARIABLES ARE DECLARED OUTSIDE ALL THE METHOD & INSIDE THE CLASS.

> <b>ACCESSIBILITY</b>:
>> * THESE VARIABLES ARE ACCESSIBLE WITHIN THE METHOD OF SAME CLASS USING CLASS NAME.
>> * THESE VARIABLES ARE ACCESSIBLE OUTSIDE THE CLASS USING CLASS NAME or CLASS OBJECT.

```python
# eg

class myclass:
    city="Hydrabad"  # Static Variable
    
    def m_add(self):
        print("Add Method")
        self.num1=int(input("Enter first number: "))  # Instance Variable
        self.num2=int(input("Enter second number: ")) # Instance Variable
        a=self.num1+self.num2 # Local Variable
        print(self.num1)
        print(self.num2)
        print(a)
        print(myclass.city) # Call city inside the method
        
        
    def m_mul(self):
        print("Multiplication Method")
        m=self.num1*self.num2 # Local Variable
        print(self.num1)
        print(self.num2)
        print(m)
        print(myclass.city) # Call city inside the method
        
obj2=myclass()
obj2.m_add()
print()
obj2.m_mul()
print()
print("Outside the class")
print("Using object: ",obj2.city)
print()
print("Using class: ",myclass.city)
```

## <font color="red">TYPES OF METHODS:</font>

### <font color="blue">a. INSTANCE METHOD:</font>
* THIS IS DEFAULT METHOD.
* A METHOD WITH `"self"` AS DEFAULT ARGUMENT IS CONSIDERED AS INSTANCE METHOD.

> __ACCESSIBILITY__:
>> * THESE METHODS CAN BE ACCESSED WITHIN THE CLASS & FROM OUTSIDE THE CLASS.<br>



>> __WITHIN CLASS__:
>>> * IT IS ACCESSED USING `"self"` KEYWORD.
>>>     * `self.methodname()`<br>
    
    
   
>> __OUTSIDE CLASS__:
>>> * IT CAN BE ACCESSED USING OBJECT.
>>>     * `object.methodname()`


__`ALL THE VARIABLES ARE ACCESSIBLE`__

```python
class myclass:
    def m_add(self):
        print("This is an Instance Method")
        
obj=myclass()
obj.m_add()
```

### <font color="blue">b. CLASS METHOD:</font>
* A METHOD WITH "cls" AS DEFAULT ARGUMENT IS CONSIDERED AS CLASS METHOD.
* THESE CLASS METHODS ARE DEFINED USING `DECORATOR / ANNOTATION`. `@classmethod`

> __ACCESSIBILTY__:
>> __OUTSIDE CLASS__:
>>    * THESE METHODS ARE ACCESSED USING TWO WAYS,
>>>  * `classname.methodname()`
>>>  * `obj.methodname()`


> __VARIABLES__:
>> * LOCAL VARIABLES
>> * STATIC VARIABLES
>>> * `classname.staticvariable`
>>> * `cls.staticvariable`

>> *  __`INSTANCE VARIABLES CAN NOT BE ACCESSED.`__
    


```python
class myclass:
    def greet(self):
        print("This is an Instance Method")
        
    @classmethod
    def method1(cls):
        print("This is from CLASS METHOD")
        

        
obj=myclass()
obj.greet()
print()
obj.method1()
print()
myclass.method1()
```

### <font color="blue">c. STATIC METHOD:</font>
* A METHOD WITH NO DEFAULT ARGUMENTS IS CALLED AS STATIC METHOD.
* THESE STATIC METHODS ARE DEFINED USING `DECORATOR / ANNOTATION`. `@staticmethod`

> __ACCESSIBILITY__:
>> __OUTSIDE CLASS__:
>>> * THESE METHOD ARE ACCESSED USING TWO WAYS,
>>>     * `classname.methodname()`
>>>     * `obj.methodname()`

> __VARIABLE__:
>> * LOCAL VARIABLES
>> * STATIC VARIABLES
>>>   * `classname.staticvariable`
>>>   * `cls.staticvariable`

>> __`INSTANCE VARIABLES CAN NOT BE ACCESSED.`__

```python
# eg

class myclass:
    def greet(self):
        print("This is an Instance Method")
        
    @classmethod
    def method1(cls):
        print("This is from CLASS METHOD")
        
    @staticmethod
    def method2():
        print("This is from Static Method")
        
        
obj=myclass()
obj.greet()
print()
obj.method1()
print()
myclass.method1()
print()
obj.method2()
print()
myclass.method2()
```

```python
# eg

class skyclass:
    city="HYDRABAD"
    
    def m1(self,n1,n2):
        self.num1=n1
        self.num2=n2
        
        a=self.num1+self.num2
        
        print(self.num1)
        print(self.num2)
        print(a)
        print(skyclass.city)
        print("***********************************")
        
    @classmethod
    def cmethod(cls):
        print("This is from Class Method")
        n1=100
        n2=200
        m=n1+n2
        print(m)
        print(skyclass.city)
        print("---------------------------------------")
        print(cls.city)
        
        
    @staticmethod
    def smethod():
        print("This is from Static Method")
        x=50
        y=90
        b=x+y
        print(b)
        print(skyclass.city)
        
        
        
        
obj1=skyclass()
obj1.m1(400,500)
print("***********************************")
obj1.cmethod()
print("**************************")
skyclass.cmethod()

obj1.smethod()
print("*****************************")
skyclass.smethod()
```

## <font color="brown">NESTED CLASS</font>

```python
class parent:
    country="INDIA"
    
    def pa1(self):
        print("This is from Parent Class Method 1")
        
    class child:
        state="ODISHA"
        
        def c1(self):
            print("This is from Child Class Method 1")
            
obj_p=parent()
print(obj_p.country)
obj_p.pa1()
```

```python
class parent:
    country="INDIA"
    
    def pa1(self):
        print("This is from Parent Class Method 1")
        
    class child:
        state="ODISHA"
        
        def c1(self):
            print("This is from Child Class Method ")
            
obj_p=parent()
print(obj_p.country)
obj_p.pa1()

obj_c=obj_p.child()
print(obj_c.state)
obj_c.c1
```

```python
class parent:
    country="INDIA"
    
    def pa1(self):
        print("This is from Parent Class Method 1")
        
        obj_c= self.child()
        print("State: ",obj_c.state)  
        obj_c.c1()
    
    class child:
        state="ODISHA"
        
        def c1(self):
            print("This is from Child Class Method ")
            
obj_p=parent()
print(obj_p.country)
obj_p.pa1()
```
