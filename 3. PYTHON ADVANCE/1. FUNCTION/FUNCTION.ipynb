## <font color="red">WHAT IS FUNCTION?</font>
* A FUNCTION IS A REPEATED BLOCK OF CODE WHICH CONSIST OF BUSINESS LOGIC.
* THIS BLOCK OF CODE CAN BE CALLED "n" NUMBER OF TIMES BASED ON THE REQUIREMENT.
> * A FUNCTION IS A BLOCK OF CODE THAT ONLY RUNS WHEN IT IS CALLED.
> * YOU CAN PASS DATA, KNOWN AS PARAMETERS, INTO A FUCTION.
> * IN SIMPLE WORD, FUNCTION IS A PARADIGM.

### <font color="brown">WHAT IS PARADIGM?</font>
* IT IS A STYLE OR WAY TO CLASSIFY PROGRAMMING LANGUAGES.
### <font color="brown">TYPES OF PARADIGM IN PYTHON:</font>
> 1. PROCEDURAL PARADIGM
> 2. FUNCTIONAL PARADIGM
> 3. MODULAR PARADIGM
> 4. OBJECT ORIENTED PROGRAMMING STRUCTURE
> 5. LOGICAL PARADIGM

## <font color="red">WHY USE FUNCTION?</font>
* NO WASTAGE OF RESOURCES.
* ENHANCEMENT IS EASY.
* DEBUGGING IS EASY.
* TO REUSE THE CODE -- DEFINE THE CODE ONCE AND USE IT ANY TIMES.

## <font color="red">TYPES OF FUNCTION:</font>
> __1. BUILT-IN FUNCTIONS:__
>    - THESE FUNCTIONS ARE PRE-DEFINED BY PYTHON.
>    - EG:-- print(), len(), input() .. 

> __2. USER DEFINED FUNCTIONS:__
>    - THESE FUNCTIONS ARE DEFINED AND USED BY THE PROGRMMER AS PER PROJECT NEEDS.
    
    
> __3. RECURSIVE FUNCTION__</br>
</br>
> __4. LAMBDA FUNCTION__

## <font color="red">COMPONENTS OF FUNCTION:</font>

### <font color="blue">a. FUNCTION DEFINATION:</font>
* __FUNCTION DECLARATION__
    * IT IS FIRST LINE OF THE DEFINITION WITH __`"def"`__ KEYWORD.
    
    
* __FUNCTION IMPLEMENTATION__
    * BLOCK OF CODE UNDER FUNCTION DEFINITION IS CALLED FUNCTION IMPLEMENTATION.

* FUNCTION DEFINITION SHOULD START WITH __`"def"`__ KEYWORD FOLLOWED BY FUNCTION NAME WHICH IDENTIFIES THE FUNCTION  DEFINITION UNIQUELY.


* FUNCTION DEFINITION CAN EXIST WITHOUT CALLING.


* FUNCTION DEFINITION SHOULD BE DEFINED BEFORE FUNCTION CALLING.


* WE CAN DEFINE MULTIPLE FUNCTION DEFINITION WITHIN THE SAME PROGRAM.


* FUNCTION DEFINTION IS EXECUTED WHEN IT IS CALLED.

### <font color="blue">b. FUNCTION CALLING:</font>
* FUNCTION NAME FOLLOWED BY PARANTHESIS IS CONSIDERED AS FUNCTION CALLING.


* FUNCTION CALLING SHOULD BE ALWAYS MENTIONED AFTER DEFINITION OF THE FUNCTION.


* FUNCTION CALLING CAN BE DONE "n" NUMBER OF TIME.


* FUNCTION CALLING CAN NOT EXIST WITHOUT FUNCTION DEFINITION, IT TRIGGERS __`ERROR`__.

```python
# FUNCTION DEFINATION
def fun_add():
    """
    THIS FUNCTION RETURNS THE ADDITION OF TWO NUMBERS.
    INPUT - ANY VALID INTEGER.
    OUTPUT - ADDITION OF TWO NUMBERS.
    """
    num1=int(input("Enter a Number: "))
    num2=int(input("Enter Second Number: "))
    a=num1+num2
    print(a)
```

```python
# FUNCTION CALL
fun_add()
```

#### <font color="blue">b. POSITIONAL ARGUMENTS:</font>
* VALUES ARE ASSIGNED TO ARGUMENT AT FUNCTION DEFINITION BASED ON POSITION IS CALLED __`POSITIONAL ARGUMENT`__.


```python
# eg:
def fun_add3(a,b,c): 
    print(a)
    print(b)
    print(c)
    add=a+b+c
    print(add)
    
fun_add3(50,60,70) 
```

#### <font color="blue">d. VARIABLE-LENGTH ARGUMENTS (*args):</font>
* IT IS A SPECIAL TYPE OF VARIABLE WHICH ACCEPT __"n"__ NUMBER OF POSITIONAL ARGUMENT.


* THESE VALUES ARE STORED IN THE FORM OF __`TUPLE`__.




```python
# eg-1
def fun_add3(*args): 
        print(args)

    
fun_add3(50,60,70,80,90,60)
```

```python
# eg-3
def multiply(*args):
    prdd=1
    
    for i in args:
        prdd=prdd * i
    
    return prdd
```

## <font color="red">TYPES OF VARIABLES:</font>

#### <font color="blue">a. LOCAL VARIABLES:</font>
* VARIABLES WHICH ARE DECLARED WITHIN A FUNCTION & CAN BE USED WITHIN THE SAME FUNCTION ONLY.


* THESE VARIABLES CAN NOT BE USED OUTSIDE OF THE DECLARED FUNCTION.

```python
# eg
def fun_add():
    
    # num1 is a variable used in this function only
    # because it is a local variable.
    
    num1=int(input("Enter a Number: "))
    num2=int(input("Enter Second Number: "))
    a=num1+num2
    print(a)
    
def fun_mul():
    
    # num1 & num2 declrae inside the fun_add function
    # they are local variable, we can not use  in another function.
    
    b=num1*num2
    print(b)
    
fun_add()
print("----------------------------------------------------")  
fun_mul()
```

<div class="alert alert-block alert-info"><b>NOTE:</b> num1 & num2 IS DEFINED IN THE fun_add() FUNCTION, SO IT IS A LOCAL VARIABLE, SO IT IS NOT USED IN fun_mul() FUNCTION.</div>

#### <font color="blue">a. GLOBAL VARIABLES:</font>
* GLOBAL VARIABLES WHICH CAN USED ANYWHERE WITHIN THE PROGRAM IS CALLED __`GLOBAL VARIABLE`__.


* GLOBAL VARIABLES ARE DECLARED IN TWO WAYS,
> * DECLARING A VARIABLE OUTSIDE ALL THE FUNCTIONS IS CALLED/CONSIDERED AS __`GLOBAL VARIABLE`__. 
> <br>
> <br>
> * A VARIABLE WHICH DECLARED WITH `"global"` KEYWORD INSIDE A FUNCTION IS ALOS CONSIDERED AS __`GLOBAL VARIABLE`__.

```python
# eg
# OUTSIDE OF ALL FUNCTION

num1=int(input("Enter a Number: "))
num2=int(input("Enter Second Number: "))

def fun_add():

    a=num1+num2
    print(a)
    
def fun_mul():

    b=num1*num2
    print(b)
    
fun_add()
print("----------------------------------------------------")  
fun_mul()
```

```python
# eg
# USING "global" KEYWORD

def fun_add():
    global num1,num2
    
    num1=int(input("Enter a Number: "))
    num2=int(input("Enter Second Number: "))
    a=num1+num2
    print(a)
    
def fun_mul():

    b=num1*num2
    print(b)
    
fun_add()
print("----------------------------------------------------")  
fun_mul()
```

### <font color="brown">NESTED FUNCTIONS:</font>
* NESTED FUNCTION CAN BE EXECUTED WITHIN THE PARENT FUNCTION.


* A FUNCTION WITHIN ANOTHER FUNCTION IS CALLED __`NESTED FUNCTION`__.


* THESE NESTED FUNCTION CAN NOT BE EXECUTED.

## <font color="dark blue">LAMBDA FUNCTION(ANONYMOUS FUNCTION):</font>
* LAMBDA FUNCTION IS CALLED ANONYMOUS FUNCTION.


* ANONYMOUS MEANS NOT IDENTIFY FOR THIS FUNCTION.


* IT IS AN ANONYMOUS INFINITE FUNCTION WHICH IS DEFINED WITH __`"lamda"`__ KEYWORD & THIS FUNCTION ACCEPTS __"n"__ NUMBER OF ARGUMENTS BUT RETURN ONLY ONE VALUE BASED ON THE EXPRESSION.

### <font color="purple">PROPERTIES OF LAMBDA FUNCTION:</font>
* IT DOESN'T HAVE ANY NAME TO IDENTIFY.


* NO `"def"` KEYWORD IS USED TO DEFINE IT.


* IT ACCEPTS `"n"` NUMBER OF ARGUMENTS.


* IT RETURN ONLY ONE VALUE WITHOUT ANY `"return"` KEYWORD.


* IT IS AN __`INLINE FUNCTION`__.


* IT IS CONSIDERED AS LIGHT WEIGHT FUNCTION.


* ITS EXECUTION IS VERY FAST COMPARED TO THAT OF TRADITIONAL FUNCTION.


* IT IS SINGLE USE FUNCTION (NO REUSABILITY.)

### <font color="purple">USE OF LAMBDA FUNCTION:</font>
* WE CAN INVOKE THE LAMBDA INTO ANOTHER PYTHON OBJECT LIKE (LIST, DICTIONARY ETC..)


* IT CAN ACT AS SOURCE OF INPUT TO HIGHER ORDER FUNCTIONS LIKE (MAP, FILTER, REDUCE)


### <font color="teal">SYNTAX</font>

### <font color="purple">DIFFERENCE BETWEEN LAMBDA vs NORMAL FUNCTION</font>
* IT DO NOT HAVE ANY NAME.


* LAMBDA HAS NO RETURN VALUE.(TECHNICALLY IT RETURNS A FUNCTION)


* LAMBDA IS WRITTEN IN ONE LINE.


* NO REUSABLE.



* __`LAMBDA FUNCTION IS USED WITH HIGHER ORDER FUNCTION.`__

### <font color="blue">a. MAP:</font>
* IT MAPS ELEMENTS OF ITERABLE WITH FUNCTION TO IMPLEMENT LOGIC.


* MAP IMPLEMENTS A BUSSINESS LOGIC TO EACH & EVERY ELEMENT OF THE ITERABLE.


* NO. OF INPUT ELEMENTS = NO. OF OUTPUT ELEMENTS.

### <font color="teal">SYNTAX</font>

```python
# eg-1

def sqr(n):  # Normal Function
    res=n**2
    return res

sqr(5)
```

```python
def sqr(n): 
    res=n**2
    return res

nlist=[9,5,8,7,4,6]

for i in nlist:
    r=sqr(i)
    print(i,end=" :")
    print(r)
```

```python
def sqr(n): 
    res=n**2
    return res

nlist=[9,5,8,7,4,6]
l=[]
for i in nlist:
    r=sqr(i)
    
    l.append(r)
print(l)
```

```python
def sqr(n): 
    res=n**2
    return res

nlist=[9,5,8,7,4,6]

m=map(sqr,nlist)
m
list(m)
```

```python
list(map(sqr,[2,3,4,5]))
```
