### INHERITANCE:
    •	It is a process of inheriting the properties of one class to another class.
    •	Class whose properties are transferred is called as Parent Class/Base Class/Super Class.
    •	Class which receives properties are called as Child Class/Derived Class/Sub Class.
    Syntax:
        Class A:     # Parent
              ------------------
              --------------------
              --------------------
        Class B(A):     # Child
            ****************
            ****************
            *****************
    	By default all user defined classes are inherited from built-in class called “object”.

### Types of Inheritance:
    1.	Simple Inheritance/Single Inheritance
        •	It is one-level of inheritance where single parent properties are inherited to single child.
    Class A:
        -------------
        -------------
    Class B(A):
        ________
        ________

    2.	Multi-Level Inheritance
        •	It is combination of multiple levels of inheritances.
        •	In this Parent properties are inherited to child class and Child class properties are inherited to Grand Child class.
        •	Grand Child can access member of Parent Class, Child Class, Grand Child Class.

        Class A:
            -----------
            ------------
            ------------
        Class B(A):
            _________
            _________
            _________
        Class C(B):
            ********
            ********
            ********

    3.	Hierarchical Inheritance
        •	In this type of inheritance, the parent class properties are inherited to multiple child class.
        •	i.e. one parent- Multiple Child
        •	This is one to many relationship.
        Class A:
            ------
            ------
        Class B(A):
            -----
            -----
        Class C(A):
            ------
            ------

### Types Of Inheritances

### 1. Single/Simple Inheritance

```python
class A:
    city="Delhi"
    def a1(self,n1,n2):
        print("This is from A1 method of Class A")
        
        self.num1=100
        self.num2=200
        
        
        a=self.num1+self.num2
        print(a)
        
        self.num1=n1
        self.num2=n2
        
class B(A):
    country="India"
    
    def b1(self,n1,n2):
        print("This is from B1 method of Class B")
        
        m=n1+n2
        print(m)
        
obj_a=A()
print(obj_a.city)
obj_a.a1(20,30)
print("________________________________")
obj_b=B()
print(obj_b.country)
obj_b.b1(10,20)
print("_________________________________")
print(obj_b.city)
obj_b.a1(20,30)
```

### 2. Multilevel Inheritance

```python
class parent:
    def p_method(self):
        print("This is from Parent Class method")
        
        for i in range(1,11):
            print(9,"*",i,"==",9*i)
            
class child(parent):
      def c_method(self):
        print("This is from Child Class method")
        
        for i in range(1,11):
            print(12,"*",i,"==",12*i)
            
class gchild(child):
      def g_method(self):
        print("This is from Grand Child Class method")
        
        for i in range(1,11):
            print(15,"*",i,"==",15*i)
            
obj_a=parent()
obj_a.p_method()

print("_____________________________________________")
print()

obj_b=child()
obj_b.c_method()
print("*******************************************")
obj_b.p_method()

print("_____________________________________________")
print()

obj_c=gchild()
obj_c.c_method()
print("*******************************************")
obj_c.p_method()
print("********************************************")
obj_c.g_method()
```
